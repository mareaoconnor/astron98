```python
import matplotlib.pyplot as plt
import numpy as np
#generate some random data
np.random.seed(0)
data1 = np.random.randint(0,101,40)
data2 = np.random.randint(0,101,40)
data3 = np.random.randint(0,101,40)
#creating figure and subplots
fig, axs = plt.subplots(2,1)
#plotting the subplots
axs[0].plot(range(len(data1)), data1, color='orange', linewidth=10, label='Data 1')
axs[0].plot(range(len(data2)), data2, color='red', linestyle='--', label='Data 2')
axs[0].set_title('Top Plot')
axs[0].legend()
#plot the bottom subplot
axs[1].scatter(range(len(data3)), data3, color='magenta', marker='D', label='Data 3')
axs[1].set_title('Bottom Plot')
axs[1].legend()
#adjusting the layout
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# constants
a = 16.02  # bar L^2 / mol^2
b = 0.1124  # L / mol
R = 0.083144  # bar L / mol K

# Pressure and volume ranges
pressure = np.linspace(1, 10, 100)  # bar
volume = np.linspace(10, 30, 100)  # L

# Create a meshgrid for pressure and volume
P, V = np.meshgrid(pressure, volume)

# Calculate temperature using the Van der Waals equation
T = (P + a / V**2) * (V - b) / R

# Create the colormap plot
plt.figure(figsize=(8, 6))
plt.imshow(T, extent=[pressure.min(), pressure.max(), volume.min(), volume.max()],
           origin='lower', aspect='auto', cmap='plasma', vmin=T.min(), vmax=T.max())
plt.colorbar(label='Temperature (K)')
plt.xlabel('Pressure (bar)')
plt.ylabel('Volume (L)')
plt.title('Temperature for Acetone (1 mole)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of random points
N_values = [int(1e3), int(1e4), int(1e5), int(1e6)]

# Function to calculate π using Monte Carlo method
def calculate_pi(N):
    points = np.random.rand(N, 2)  # Generate random points in 1x1 square
    distance = np.sqrt(points[:, 0]**2 + points[:, 1]**2)  # Calculate distance from origin
    inside_circle = distance[distance <= 1]  # Points within distance 1 from origin
    pi_approx = 4 * len(inside_circle) / N  # Approximation of π
    return pi_approx

# Calculate π for different N values
pi_values = []
for N in N_values:
    pi_values.append(calculate_pi(N))
    print(f"N = {N}: π ≈ {pi_values[-1]}")

# Create plot for N = 1e4
N = int(1e4)
points = np.random.rand(N, 2)
distance = np.sqrt(points[:, 0]**2 + points[:, 1]**2)
inside_circle = points[distance <= 1]
outside_circle = points[distance > 1]

plt.figure(figsize=(6, 6))
plt.scatter(inside_circle[:, 0], inside_circle[:, 1], color='blue', label='Inside Circle')
plt.scatter(outside_circle[:, 0], outside_circle[:, 1], color='red', label='Outside Circle')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Monte Carlo Estimation of π (N = 1e4)')
plt.axis('equal')
plt.text(0.1, 0.9, f'π ≈ {pi_values[1]}', transform=plt.gca().transAxes, fontsize=12, color='black')
plt.legend()
plt.show()
```
